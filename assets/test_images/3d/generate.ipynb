```python
import numpy as np
from pathlib import Path
```

```python
GRAYSCALE_IMAGES_PATH = Path(".")
```

```python
image_0 = np.array(range(0, 256, 32)).reshape((2, 2, 2))
np.save(GRAYSCALE_IMAGES_PATH / "image_0_2x2x2.npy", image_0)
print(image_0)
```

```python
np.random.seed(42)
image_1= np.random.randint(0, 256, size=(2, 2, 2), dtype=np.uint8)
np.save(GRAYSCALE_IMAGES_PATH / "image_1_2x2x2.npy", image_1)
print(image_1)
```
