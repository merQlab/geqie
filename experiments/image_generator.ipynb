```python
n_images_to_generate = 16
grayscale_image_dim = (8, 8)
rgb_image_dim = (8, 8, 3)
seed = 71

import os
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(seed)
current_location = os.path.dirname(os.path.abspath("__file__"))
path_to_save_generated_images_grayscale = current_location + "/generated_images/grayscale/"
path_to_save_generated_images_rgb = current_location + "/generated_images/rgb/"
os.makedirs(path_to_save_generated_images_rgb, exist_ok=True)
os.makedirs(path_to_save_generated_images_grayscale, exist_ok=True)

for i in range(n_images_to_generate):
	img = np.random.randint(0, 256, size=grayscale_image_dim, dtype=np.uint8)
	plt.imsave(path_to_save_generated_images_grayscale + f"image_{i:03d}.png", img, cmap="gray")
	
	rgb_img = np.random.randint(0, 256, size=rgb_image_dim, dtype=np.uint8)
	plt.imsave(path_to_save_generated_images_rgb + f"image_{i:03d}.png", rgb_img)

```
