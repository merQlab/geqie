```python
import os
import matplotlib.pyplot as plt
import numpy as np

from itertools import product
from pathlib import Path

from tqdm import tqdm
```

```python
N_IMAGES_TO_GENERATE = 8
GRAYSCALE_IMAGE_DIMS = [(2, 2), (4, 4), (8, 8)]
RGB_IMAGE_DIMS = [(2, 2, 3), (4, 4, 3), (8, 8, 3)]
SEED = 71

np.random.seed(SEED)
```

```python
CURRENT_PATH = Path(os.path.abspath("."))

GRAYSCALE_IMAGES_PATH = CURRENT_PATH / "grayscale"
RGB_IMAGES_PATH = CURRENT_PATH / "rgb"
IMAGE_NAME_TEMPLATE = "image_{index:03d}_{w}x{h}_.png"

GRAYSCALE_IMAGES_PATH.mkdir(parents=True, exist_ok=True)
RGB_IMAGES_PATH.mkdir(parents=True, exist_ok=True)


for i, (image_dim, _) in tqdm(list(enumerate(product(GRAYSCALE_IMAGE_DIMS, range(N_IMAGES_TO_GENERATE))))):
    w, h = image_dim
    img = np.random.randint(0, 256, size=image_dim, dtype=np.uint8)
    plt.imsave(GRAYSCALE_IMAGES_PATH / IMAGE_NAME_TEMPLATE.format(index=i, w=w, h=h), img, cmap="gray")

for i, (image_dim, _) in tqdm(list(enumerate(product(RGB_IMAGE_DIMS, range(N_IMAGES_TO_GENERATE))))):
    w, h, rgb = image_dim
    img = np.random.randint(0, 256, size=image_dim, dtype=np.uint8)
    plt.imsave(RGB_IMAGES_PATH / IMAGE_NAME_TEMPLATE.format(index=i, w=w, h=h), img, cmap="gray")
```
